```python
# Importing pandas as it seems to be missing

# Part 1: Find the best test accuracy for each model_name and fold, along with the corresponding epoch and learning rate
best_accuracy_per_fold = df.groupby(['model_name', 'fold']).apply(
    lambda x: x.loc[x['test_accuracy'].idxmax(), ['epoch', 'lr', 'test_accuracy']]
).reset_index()

# Part 2: For each model_name and learning rate, average the best epoch across all folds
average_best_epoch_per_lr = df.groupby(['model_name', 'lr', 'fold']).apply(
    lambda x: x.loc[x['test_accuracy'].idxmax(), 'epoch']
).groupby(['model_name', 'lr']).mean().reset_index()

# # Display the results to the user
# import ace_tools as tools; tools.display_dataframe_to_user(name="Best Accuracy Per Fold", dataframe=best_accuracy_per_fold)
# import ace_tools as tools; tools.display_dataframe_to_user(name="Average Best Epoch Per LR", dataframe=average_best_epoch_per_lr)

# best_accuracy_per_fold, average_best_epoch_per_lr
```
